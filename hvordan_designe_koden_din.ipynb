## Iterativ kodedesign

Prinsippet vi følger her er at i steden for å prøve å skrive kode rett ut dersom man er mer eller mindre clueless på hvordan man skal løse oppgaven, så starter vi med å beskrive en veldig generell beskrivelse av hva løsningen må gjøre i ren, menneskelig tekst, og deretter gjøre beskrivelsen om til en steg-liste som vi gradvis gjør mer og mer "kodete" til det faktisk __er__ en python-kode som kan kjøre. 

Dette gjør det mye enklere å komme i gang, fordi du som oftest klarer å beskrive hva som skal til for å løse oppgaven på et generellt nivå. Når du først har etablert den generelle flyten så er det som regel mye enklere å gjøre den gradvis mer og mer presis og kode-lignende enn å bare starte med å prøve seg fram med å kode noe direkte.

Det er altså flere fordeler med å gjøre det slik:
1. Det gjør det enklere å planlegge hvordan du skal løse et sammensatt problem
2. Det minimerer sansynligheten for at du får feil/bugs i slutt-koden din
3. Dersom du likevel står fast så er det ikke uvanlig at sensor faktisk gir deg litt uttelling dersom du i det minste har skrevet ned hva du _ville_ ha gjort steg for steg.

### Steg-liste:
---

#### Forarbeid: Først, les oppgaven nøye og vær sikker på at du forstår _hva_ programmet ditt skal gjøre på et overordnet nivå
  * Finn ut om du skal f.eks. lage et spill av noe slag, gjøre noe med filer, noe matematiske greier, osv.
#### Iterasjon 1: Skriv ned med dine egne ord hva programmet ditt må gjøre på et generelt nivå
  * Her skal du ikke skrive kode, men vanlig, menneskelig, tekst.
  * Start med å bestemme deg for hva slags faser programmet ditt består av (f.eks. 1: les fil, 2: sammenlign med et brukerinput, 3: skriv til fil)
  * Skriv en generell setning eller to for hver av disse fasene som forklarer hva som må gjøres.
#### Iterasjon 2: Skriv ned litt mer konkret hva som må gjøres i hver av fasene du identifiserte i iterasjon 1
  * Først, prøv å avgjøre hva slags datastrukturer du trenger for de forskjellige fasene
    * Dette kan f.eks. være strenger, lister, liste med lister (2D liste), dictionaries, osv.
    * Finn ut hvis noen av disse må være globale (altså at de trengs overalt og gjennom hele programmet)
  * Deretter, prøv å avgjøre hvorvidt du vil trenge if-else, while-løkker, for-løkker, o.l. for de forskjellige fasene
  * Skriv en kort setning for hvert steg i de forskjellige fasene dine.
    * Bruk fremdeles egne ord, ikke kode
    * Eks: "Fase 1: Først må jeg åpne filen, så må jeg lese hver linje i filen, og hver linje skal bli en liste som puttes inn i en større liste."
    * Tips: Dersom du ser at du bruker ord som 1: "hver/alle/...ganger", 2: "hvis/dersom/bare", så tyder dette på at du vil få bruk for 1: løkker, og 2: if-else
### Iterasjon 3: Skriv hver fase som pseudokode
  * Skriv ikke faktisk koden _helt_ ennå. For å være sikker på at selve kodingen blir enkel på slutten koster vi på oss 1 eller flere slike iterasjoner som iterasjon 3.
  * Prinsippet er at man skal gjøre den "menneske-nivå" beskrivelsen av programmet ditt gradvis mer og mer "kodete" for hver iterasjon du gjør.
  * Skriv ned hva som skal skje på hver linje i programmet ditt, men bruk fremdeles menneske-ord:
    * "Linje a: Les neste linje i filen, Linje b: Gjør om strengen du leser om til en liste hvor du isolerer verdiene som er adskilt med komma i strengen"
### Iterasjon 4: Skriv den faktiske koden
  * Nå har du faktisk kommet så langt at du har laget et detaljert step-by-step guide til hvordan koden din skal funke!
  * Eneste som skal til nå er å bytte ut de forholdsvis enkle stegene du har laget med en faktisk kodelinje som gjør det som skal skje i det steget.
### Avslutning: Se over den ferdige koden, test den og gjør finpuss


### Eksempel:
---

#### Oppgavebeskrivelse

Du er kjempeletta etter du er ferdig med den siste eksamenen din, og i et øyeblikks galskap bestemte du deg for å invitere hele klassa di på julebord hjemme i kollektivet ditt. Etter du har sendt ut invitasjonene så finner du ut at alle i klassa di har fått for seg at de skal ta den __fullstendig__ ut på julebordet, og du er derfor redd for at det skal bli skader på ting i kollektivet ditt.

Løsningen din er å ta i bruk ting du har lært i ITGK (for det er jo normalt i en slik situasjon...)! Det du trenger å gjøre er å lage et program som lar deg registrere hvem som har ødelagt hva i løpet av julebordet ditt. Hver person som ødelegger noe skal registreres med følgende info:
* Fornavn: (f.eks. Håvard)
* Beskrivelse av hva de ødela: (f.eks. Sølte vin i sofaen)

Til slutt så skal du skrive ut alle registrerte hendelser til en fil, slik at du kan lagre den til dagen derpå når du skal begynne å sende ut vippskrav.

<br/><br/><br/>

#### Designprosess, steg for steg

__Iterasjon 1__:
Programmet består av 3 faser. I fase 1 skal vi gjøre det mulig for brukeren å si i fra om hen vil registrere en ny skade eller om julebordet er ferdig. Fase 2 skal skje hver gang en ny person skal registreres. Fase 3 skal skje når brukeren har sagt at julebordet er ferdig, og da skal vi skrive ned alle registrerte personer og hva de skadet til en fil.  

__Iterasjon 2__:
Datastrukturer som jeg tror jeg trenger: Dictionaries for å represente en registrering, global liste som inneholder alle registreringene.
Andre ting som jeg trenger: If-else for å sjekke om vi skal registrere eller om julebordet er over, while-løkke som lar oss fortsette å registrere til julebordet er over.

* Fase 1: Avgjør om brukeren vil registrere eller avslutte. Dersom vi skal registrere, gå til fase 2, og hvis ikke, gå til fase 3. 
* Fase 2: Spør om navnet på personen og beskrivelse av hva som ble ødelagt. Putt svarene inn i en dictionary. Putt dictionary i lista med alle registrerte skader. Gå tilbake til fase 1. 
* Fase 3: Åpne en fil for skriving. Skriv navn og beskrivelse for hver registerte hendelse på en egen linje i filen, adskilt med f.eks. et semikolon (;). Lukk filen og avslutt programet.

__Iterasjon 3__:

Globale variabler:
1. registreringer = tom liste

Fase 1:
1. julebord_fortsetter = True
2. så lenge julebord_fortsetter:
   * neste_handling = input fra brukeren
   * dersom neste_handling er å registrere:
     * gå til fase 2
   * ellers:
     * gå til fase 3
     * jule_bord_fortsetter = False
3. si "takk for nå og god jul"

Fase 2:
1. navn = input fra brukeren
2. beskrivelse = input fra brukeren
3. lag dictionary med navn og beskrivelse
4. putt dictionary i lista med registreringer

Fase 3:
1. åpne fil
2. for hver registrering i lista over registreringer:
   * lag en streng på formatet <code>navn</code>;<code>beskrivelse</code>
   * skriv denne strengen til filen og skift linje
3. lukk fil

__Iterasjon 4__:

```python
registreringer = []

# Funksjon for fase 2
def registrer():
    navn = input('Navnet til personen du ønsker å registrere: ').strip()
    beskrivelse = input('Beskrivelse av skaden: ').strip()
    ny_hendelse = {
        'navn': navn,
        'beskrivelse': beskrivelse
    }
    registreringer.append(ny_hendelse)

# Funksjon for fase 3
def skriv_til_fil():
    filnavn = './julebord_2025.txt'
    fil = open(filnavn, 'w')
    for hendelse in registreringer:
        linje = hendelse['navn'] + ';' + hendelse['beskrivelse'] + '\n'
        fil.write(linje)
    fil.close()


# Resten er fase 1
julebord_fortsetter = True

while (julebord_fortsetter):
    neste_handling = input('Hva skal gjøres nå (registrer/avslutt):').lower().strip()
    
    if neste_handling == 'registrer':
        registrer()
    elif neste_handling == 'avslutt':
        skriv_til_fil()
        julebord_fortsetter = False
    else:
        print('Ugyldig alternativ. Skriv enten "registrer" eller "avslutt".')

print('Takk for nå og god jul!')
```
