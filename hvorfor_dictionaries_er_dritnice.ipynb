### Spørsmål 13 (stjernemerket)

Din venn Ingrid jobber som studieveileder ved NTNU. Hun må ofte slå opp kompliserte regler for
å avgjøre om studenter får gå opp til eksamen i ulike emner. Dette er tidkrevende, så hun har
bedt deg om å lage et program som kan automatisere vurderingen. I første omgang ønsker hun
en enkel versjon som bare inkluderer tre emner. <br/>
Parametere som skal brukes i programmet (Se nederst for ferdig utgitt kode):
- Emnekode (Alternativer: {TDT4111, TMA4400, MET1001})
- Godkjent øvingsopplegg fra før (Alternativer: {ja, nei})
- Antall godkjente innleveringer
- Tatt for plagiat (Alternativer: {ja, nei})
- Antall ganger studenten har tatt eksamen tidligere
- Legeerklæring (Alternativer: {ja, nei})

Generell regel ved NTNU:
- En student får vanligvis maksimalt 3 forsøk på samme eksamen. Dersom denne grensen
er overskredet, gis alltid avslag.
- Et unntak gjelder for TMA4400, hvor det er lov med inntil 4 forsøk.
- Dersom studenten er tatt for plagiat skal det alltid gis avslag, uansett andre forhold

Følgende krav gjelder for de tre emnene:
- TDT4111:
  - Minst 8 av 10 øvinger må være godkjent
  - Dersom studenten har legeerklæring, holder det med minst 5 øvinger
  - Hvis øvingsopplegget allerede er godkjent fra før har studenten uansett adgang

- TMA4400:
  - Minst 6 av 10 øvinger må være godkjent
  - Med legeerklæring er kravet minst 4 øvinger
  - Øvingsopplegget kan også være forhåndsgodkjent

- MET1001:
  - Minst 5 av 8 øvinger må være godkjent
  - Med legeerklæring er kravet minst 3 øvinger
  - Øvingsopplegget kan også være forhåndsgodkjent

Programmet skal skrive ut enten "Adgang" eller "Avslag" basert på reglene over.

<br/><br/>

## Her er hvordan løsningsforslaget foreslo at man løste denne:

````python
emne = input("Emnekode: ")                                                               # {TDT4111, TMA4400, MET1001}
godkjent_ovingsopplegg = input("Har eleven godkjent øvingsopplegg fra før (ja/nei): ")   # {ja, nei}
innlev = int(input("Antall godkjente innleveringer: "))                                  # Antall godkjente innleveringer
plagiat = input("Har eleven blitt tatt for plagiat (ja/nei): ")                          # {ja, nei}
antall_forsok = int(input("Antall ganger studenten har tatt eksamen tidligere: "))       # Antall ganger studenten har tatt eksamen tidligere
lege = input("Har eleven legeærklæring (ja/nei): ")                                      # {ja, nei}

if plagiat == "ja":
    print("Avslag")
elif emne == "TDT4111":
    if (godkjent_ovingsopplegg == "ja" or innlev >= 8 or (lege == "ja" and innlev >= 5)) and antall_forsok < 3:
        print("Adgang")
    else:
        print("Avslag")
elif emne == "TMA4400":
    if (godkjent_ovingsopplegg == "ja" or innlev >= 6 or (lege == "ja" and innlev >= 4)) and antall_forsok < 4:
        print("Adgang")
    else:
        print("Avslag")
elif emne == "MET1001":
    if (godkjent_ovingsopplegg == "ja" or innlev >= 5 or (lege == "ja" and innlev >= 3)) and antall_forsok < 3:
        print("Adgang")
    else:
        print("Avslag")
````

## Det er en helt fair måte å løse oppgaven på, men her er en løsning som jeg syns er bedre:

````python
EMNER = {
    'TDT4111': {
        'max_forsok': 3,
        'min_godkjente_ovinger': 8,
        'min_godkjente_ovinger_med_legeerklaering': 5,
        'godkjennes_tidligere_ovinger': True
    },
    'TMA4400': {
        'max_forsok': 4,
        'min_godkjente_ovinger': 6,
        'min_godkjente_ovinger_med_legeerklaering':4,
        'godkjennes_tidligere_ovinger': True
    },
    'MET1001': {
        'max_forsok': 3,
        'min_godkjente_ovinger': 5,
        'min_godkjente_ovinger_med_legeerklaering': 3,
        'godkjennes_tidligere_ovinger': True
    }
}

def tilgang_paa_eksamen(soknad):
    if soknad['emne'] not in EMNER:
        print(f'Fant ikke emnekrav for emne med emnekode: {soknad["emne"]}.')
        return False

    emnekrav = EMNER[soknad['emne']]

    if soknad['tatt_i_plagiat'] or soknad['antall_forsok'] > emnekrav['max_forsok']:
        return False
    elif emnekrav['godkjennes_tidligere_ovinger'] and soknad['godkjent_ovingsopplegg']:
        return True

    if soknad['har_legeerklaering']:
        return soknad['antall_godkjente_innleveringer'] >= emnekrav['min_godkjente_ovinger_med_legeerklaering']
    else:
        return soknad['antall_godkjente_innleveringer'] >= emnekrav['min_godkjente_ovinger']

soknad = {
    'emne': input("Emnekode: ").upper(), # {TDT4111, TMA4400, MET1001}
    'godkjent_ovingsopplegg': input("Har eleven godkjent øvingsopplegg fra før (ja/nei): ").lower() == 'ja', # {ja, nei}
    'antall_godkjente_innleveringer': int(input("Antall godkjente innleveringer: ")),
    'tatt_i_plagiat': input("Har eleven blitt tatt for plagiat (ja/nei): ").lower() == 'ja', # {ja, nei}
    'antall_forsok': int(input("Antall ganger studenten har tatt eksamen tidligere: ")),
    'har_legeerklaering': input("Har eleven legeærklæring (ja/nei): ").lower() == 'ja' # {ja, nei}
}

print('Adgang' if tilgang_paa_eksamen(soknad) else 'Avslag')
````

Grunnen til at jeg syns at denne løsningen er bedre er fordi at denne er mye enklere å utvide enn varianten fra løsningsforslaget. Se for deg om det hadde vært 10 forskjellige emner man skulle sjekke for!

* Hva som hadde skjedd med koden i denne alternative løsningen:
  * Vi kan lagre de forskjellige regelsettene i separate filer og lese inn den vi vil bruke!


### Eksempel på hvordan du kan utvide den siste løsningen til å lese regelsettet fra en fil istedet

Si at vi har overført innholdet i <code>EMNER</code> i forskjellige txt-filer hvor selve reglene er litt forskjellige:
* _regelsett_1.txt_
* _regelsett_2.txt_
* _regelsett_3.txt_
* osv..

Innhold i _regelsett_1.txt_:

```txt
TDT4111, 3, 8, 5, True
TMA4400, 4, 6, 4, True
MET1001, 3, 5, 3, True
```

Innhold i _regelsett_2.txt_:

```txt
TDT4111, 3, 6, 4, True
TMA4400, 2, 6, 4, False
MET1001, 3, 4, 1, True
```

Da kan vi skrive om koden slik som dette:

````python
EMNER = {}

def hent_regler(versjon=1):
    global EMNER
    filnavn = f'./regelsett_{versjon}.txt'
    with open(filnavn, 'r') as fil:
        for linje in fil:
            emne = linje.split(',')
            emnekode = emne[0]
            EMNER[emnekode] = {
                'max_forsok': int(emne[1]),
                'min_godkjente_ovinger': int(emne[2]),
                'min_godkjente_ovinger_med_legeerklaering': int(emne[3]),
                'godkjennes_tidligere_ovinger': bool(emne[4])
            }

versjon = int(input("Hvilke versjon av regelsettet skal brukes: ")) # {1, 2, 3, ...}
hent_regler(versjon)

# Resten av koden er helt lik som i sta ...
````

Da kan vi gjøre dette her ENDA enklere! Bare se her:

````python
import json

EMNER = {}

def hent_regler(versjon=1):
    global EMNER
    filnavn = f'./regelsett_{versjon}.json'
    with open(filnavn, 'r') as fil:
        EMNER = json.load(fil)


versjon = int(input("Hvilke versjon av regelsettet skal brukes: ")) # {1, 2, 3, ...}
hent_regler(versjon)

# Resten av koden er helt lik som i sta ...
````
